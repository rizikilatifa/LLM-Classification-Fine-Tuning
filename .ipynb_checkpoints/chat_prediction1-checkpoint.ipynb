```python
#Validate winner encoding
winner_cols = ['winner_model_a', 'winner_model_b', 'winner_tie']

data['winner_sum'] =data[winner_cols].sum(axis=1)

data[data['winner_sum']!=1]
```

```python
# checking for class imbalance
data[winner_cols].sum()
```

```python
data['preferred_model'] = np.where(
    data['winner_model_a'] == 1, data['model_a'],
    np.where(data['winner_model_b'] == 1, data['model_b'], 'tie')
)
```

```python
data[data['response_a'].isna() | data['response_b'].isna()]
```

```python
# count total comparisons per model
comparisons = (data['model_a'].value_counts()
               .add(data['model_b'].value_counts(), fill_value = 0)
              )
```

```python
#(how many times was a specific model evalusted)
comparisons
```

```python
# count wins per model
wins =data.loc[data['preferred_model'] != 'tie', 'preferred_model'].value_counts()
```

```python
wins
```

```python
win_rates = (
    pd.DataFrame({
        'wins': wins,
        'comparisons': comparisons
    })
    .fillna(0)
)

win_rates['win_rate'] = win_rates['wins'] / win_rates['comparisons']
```

```python
win_rates = win_rates.sort_values('win_rate', ascending=False)
win_rates
```

```python
#Building a clean winner-loser table

wins_df =data[data['winner_tie'] != 1].copy()

wins_df['winner'] = np.where(
    wins_df['winner_model_a'] == 1,
    wins_df['model_a'],
    wins_df['model_b']
)

wins_df['loser'] = np.where(
    wins_df['winner_model_a'] == 1,
    wins_df['model_b'],
    wins_df['model_a']
)
```

```python
wins_df
```

```python
# create pairwise win matrix
pairwise_matrix = pd.crosstab(
    wins_df['winner'],
    wins_df['loser']
)
```

```python
pairwise_matrix
```

```python

np.fill_diagonal(pairwise_matrix.values, 0)
```

```python
pairwise_matrix
```

```python
import  matplotlib.pyplot as plt
import seaborn as sn

plt.figure(figsize= (12,10))

sn.heatmap(
    pairwise_matrix,
    annot = True,
    fmt = "d",
    cmap="Blues",
    linewidth= 0.5
)

plt.title("Pairwise Win Matrix(Win Counts)")
plt.xlabel("loser")
plt.ylabel("winner")
plt.show()
                     
```
